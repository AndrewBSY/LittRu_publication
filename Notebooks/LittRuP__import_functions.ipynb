```python

def titres(
    fig,
    ax,
    t1="", y1=35,
    t2="", y2=20,
    t3="", y3=7
):

    fig.canvas.draw()

    bbox = ax.get_position()
    xcentre = (bbox.x0 + bbox.x1) / 2

    if t1:
        trans = offset_copy(
            fig.transFigure,
            fig=fig,
            x=0,
            y=y1,
            units="points"
        )

        fig.text(
            xcentre,
            bbox.y1,
            t1,
            transform=trans,
            ha="center",
            fontsize=14,
            fontweight="bold"
        )

    if t2:
        trans = offset_copy(
            fig.transFigure,
            fig=fig,
            x=0,
            y=y2,
            units="points"
        )

        fig.text(
            xcentre,
            bbox.y1,
            t2,
            transform=trans,
            ha="center",
            fontsize=12,
            style="italic"
        )

    if t3:
        trans = offset_copy(
            fig.transFigure,
            fig=fig,
            x=0,
            y=y3,
            units="points"
        )

        fig.text(
            xcentre,
            bbox.y1,
            t3,
            transform=trans,
            ha="center",
            fontsize=10
        )

```
